Importing the library.

```python
import numpy as np
import pandas as pd
```

Importing the dataset.

```python
dataset = pd.read_csv("D:\MLprojects\champions.csv")
```

The dataset is the statistics of all of 147 champions in League of Legends from 27/11/2019 to 27/12/2019.

```python
dataset
```

The model will predict the average amount of gold a random champion gained in a game based on Win Rate, Games Played, KD/A, and CS in any time periods.

Divide the dataset into independent variables and dependent variables.

```python
X = dataset.iloc[:, 1:5].values
x = pd.DataFrame(X, columns=["Win Rate", "Games Played", "KD/A", "CS"])
```

```python
x
```

```python
Y = dataset.iloc[:, 5].values
y = pd.DataFrame(Y, columns=["Gold"])
```

```python
y
```

Inspecting which independent variable have the least influence on the dependent variable using backward elimination. Choose the significant level for this model to be SL = 0.05.

```python
import statsmodels.api as sm
x_model = X[: , [0,1,2,3]]
regressor_OLS = sm.OLS(Y,x_model).fit()
regressor_OLS.summary()
```

The predictor x2 (Games Played) has the highest p-value, which is 0.322. This p-value is greater than the significant level (SL = 0.05), so remove x2 (Games Played) from the model.

```python
x_model = X[: , [0,2,3]]
regressor_OLS = sm.OLS(Y,x_model).fit()
regressor_OLS.summary()
```

The predictor x3 (CS) has the highest p-value, which is 0.316. This p-value is greater than the significant level (SL = 0.05), so remove x3 (CS) from the model.

```python
x_model = X[: , [0,2]]
regressor_OLS = sm.OLS(Y,x_model).fit()
regressor_OLS.summary()
```

Two remaining predictors, which are Win Rate and KD/A, have the p-values that are lower than the significant value (SL = 0.05). Moreover, the adjusted R-squared value nearly remains the same (0.989) through the process of backward elimination. Therefore, choose the two independent variables, Win Rate and KD/A, to fit into the model.

```python
X_input = X[: , [0,2]]
```

```python
X_input
```

```python
x_input = pd.DataFrame(X_input, columns=["Win Rate", "KD/A"])
```

Splitting the dataset into the Training set and Test set.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_input, Y, test_size = 0.2, random_state = 13)
```

Fitting Multiple Linear Regression to the Training set.

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

Predicting the Test set result.

```python
Y_pred = regressor.predict(X_test)
```

Compare the predicted gold values to the real gold values to see the efficiency of the model.

```python
Y_compare = np.array([Y_test, Y_pred]).transpose()
Y_compare_table = pd.DataFrame(Y_compare, columns=["Real Gold Value", "Predicted Gold Value"])
```

```python
Y_compare_table
```

From the result table, it can be seen that the model is efficient at most points. The deviation of predicted gold value is lower than 5% compared to the real gold value. However, for some points, the model predicts the result with deviation more than 5%. For example, at index 26, the model predicts 9820 compared to the real value 7598. The deviation is nearly 30%.
